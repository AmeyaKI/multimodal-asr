```python
from datasets import load_dataset
from IPython.display import Audio as ipyAudio
import torchcodec, torchaudio
import matplotlib.pyplot as plt
import numpy as np
```

```python
from google.colab import drive, files
drive.mount('/content/drive')
```

```python
file_path = 'clean/test-00001-of-00009.parquet' # downloading 1 parquet file
dataset = load_dataset("MLCommons/peoples_speech", data_files=file_path) # downloading one specific file
```

```python
sample = next(iter(dataset['train'])) # sampling first row of data
audio = sample['audio']['array']
sample_rate = sample['audio']['sampling_rate']
real_text = sample['text']
```

```python
# Graphical representation

time_ms = np.arange(len(audio)) * (1000 / sample_rate)

plt.figure(figsize=(15,5))
plt.plot(time_ms, audio)
plt.xlabel('Time (ms)')
plt.ylabel('Amplitude')
```

```python
# creating audio file
audio_file = ipyAudio(audio, rate=sample_rate)
audio_file
```

```python
# Saving audio file to drive
import os
save_file_path = '/content/drive/MyDrive/Colab Notebooks/data/asr'
file_name = 'output_audio'
with open(f'{save_file_path}/{file_name}.wav', 'wb') as f:
    f.write(audio_file.data)
```

```python
# saving audio file text to drive
with open(f'{save_file_path}/{file_name}.txt', 'w') as f:
    f.write(real_text)
```

```python
# verifying audio + txt files were saved
os.listdir(save_file_path)
```
